```python
'''6. Take an integer input 
    - if number is divisible by 3 -> print Divisible by 3 
    - else print - not divisble by 3 '''

num = int(input("enter number:"))
if num%3==0 and num>0:
    print(f"{num} is divisible by 3")
else:
    print(f"{num} is not divisible by 3")
```

```python
'''7. Take a number 
    - if number is between 1 to 100 
        - print Even or Odd 
    - else print Out of range '''

num=int(input("enter number:"))
if 1<num<100:
    print("number is between 1 to 100 range")
    if num%2==0:
        print("number is even")
    else:
        print("number is odd")
else:
    print("number is Out of range")

```

```python
'''10. Input age and qualification 
    -if age >=18 
        - if qualification is 12th pass -> Eligible 
    - else - Not qualified'''

age =int(input("enter age:"))
qualification=input("enter qualification")
if age>=18:
    if qualification=='12th pass':
        print("Eligible")
    else:
        print("not eligible")
else:
    print("age is less than 18")
```

```python
'''11.  Input balance and withrowal amount 
    - if balance > 0 
        - if withrowal <= balance - > Withrowal successful 
        - else "Insufficient balance"
    - Else - Account - Empty '''

bal=int(input("enter balance:"))
amt=int(input("enter withrowal amount:"))
if bal>0:
    if amt<=bal:
        print("Withrowal successful")
    else:
        print("Insufficient balance")
else:
    print("Account - Empty")

```
