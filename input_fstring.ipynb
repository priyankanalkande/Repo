```python
#8. Take a amount and print: 
    Final amount after discount: <value>
    Rule:
        - >= 5000 -> 20% discount 
        - >= 2000 -> 10% 
        - else - No discount 
        #price - (price * discount / 100)
```

```python
amt=int(input("Enter amount:"))
discounted = (amt>=5000)*20 +(amt>=2000 and amt<=2000)*10
final_amt =amt - (amt * discounted / 100)
print(f"Final amount after discount: {final_amt}")
```

```python
amt = float(input("Enter amount: "))
discount = 0

if amt >= 5000:
    discount = 20
elif amt >= 2000:
    discount = 10

final_amt = amt - (amt * discount / 100)
print(f"Final amount after discount: {final_amt}")
```
